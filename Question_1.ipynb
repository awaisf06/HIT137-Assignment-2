```python
import spacy
from transformers import AutoModelForTokenClassification, AutoTokenizer

# Load text
with open("alltexts.txt", "r") as f:
    text = f.read()

# Load spaCy models (adjust max_length if needed)
nlp_sci = spacy.load("en_core_sci_sm")
nlp_sci.max_length = 2000000  # Adjust based on available RAM
nlp_bc5cdr = spacy.load("en_ner_bc5cdr_md")
nlp_bc5cdr.max_length = 2000000  # Adjust based on available RAM

```

```python
# Process text in chunks (recommended)
chunk_size = 100000  # Adjust as needed
all_diseases_sci = []
all_drugs_sci = []
chunk_size = 100000
for i in range(0, len(text), chunk_size):
    chunk = text[i:i+chunk_size]  # Extract chunk by slicing
    diseases_sci = []
    drugs_sci = []
    for doc in nlp_sci.pipe([chunk], batch_size=50):
        for ent in doc.ents:
            if ent.label_ == "DISEASE":
                diseases_sci.append(ent.text)
            elif ent.label_ == "CHEMICAL":  # Assuming CHEMICAL for drugs
                drugs_sci.append(ent.text)
    all_diseases_sci.extend(diseases_sci)
    all_drugs_sci.extend(drugs_sci)


# Combine results from all chunks
diseases_sci = all_diseases_sci
drugs_sci = all_drugs_sci

```
